```python
import openai
import requests
```

```python
ai_api_key = ''
```

```python
def ai_chats(messages):
    
    headers = {
        
        'Content-Type': 'application/json',
        'Authorization': f'Bearer {ai_api_key}'
    }
    data={
        'messages': messages

    }
    response = requests.post('https://api.openai.com/v1/chat/completions', headers=headers, json=data)
    response_data = response.json()

    choices = response_data.get('choices')
    if choices and len(choices) > 0:
        return choices[0].get('message', {}).get('content', '')
    return ''


```

```python
def count_tokens(text):
    # A simple token count function
    return len(text.split())
```

```python
def truncate_conversation(conversation, max_tokens):
    conversation_copy = conversation.copy()
    total_tokens = sum(count_tokens(msg['content']) for msg in conversation_copy)
    
    # Keep removing user messages until the conversation fits within the token limit
    while total_tokens > max_tokens:
        if len(conversation_copy) == 2:
            break  # To ensure at least one system message is kept
        removed_msg = conversation_copy.pop(1)  # Remove the last user message
        total_tokens -= count_tokens(removed_msg['content'])
    
    return conversation_copy
```

```python
def main():
    conversation = [{'role': 'system', 'content': 'You are AI model 1.'},
                    {'role': 'system', 'content': 'You are AI model 2.'}]

    while True:
        # AI Model 1's turn
        print("Conversation (AI Model 1):", conversation)
        truncated_conversation = truncate_conversation(conversation, max_tokens=4096)
        reply_model1 = ai_chats(truncated_conversation + [{'role': 'AI Model 1', 'content': ''}])
        print("AI Model 1: ", reply_model1)
        conversation.append({'role': 'AI Model 1', 'content': reply_model1})

        # Exit loop if AI Model 1's response contains 'exit'
        if 'exit' in reply_model1.lower():
            break

        # AI Model 2's turn
        print("Conversation (AI Model 2):", conversation)
        truncated_conversation = truncate_conversation(conversation, max_tokens=4096)
        reply_model2 = ai_chats(truncated_conversation + [{'role': 'AI Model 2', 'content': ''}])
        print("AI Model 2: ", reply_model2)
        conversation.append({'role': 'AI Model 2', 'content': reply_model2})

        # Exit loop if AI Model 2's response contains 'exit'
        if 'exit' in reply_model2.lower():
            break
```

```python
if __name__ == "__main__":
    main()
```
